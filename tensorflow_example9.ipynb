```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = tf.image.decode_jpeg(tf.read_file("data/1-content.jpg"))
```

```python
init_op = tf.global_variables_initializer()
with tf.Session() as s:
    s.run(init_op)
    img = s.run(image)
    print(img.shape)
    plt.figure()
    plt.imshow(img)
    plt.show()
```
