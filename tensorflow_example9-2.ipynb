```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = tf.image.decode_jpeg(tf.read_file("data/1-content.jpg"))
```

```python
resized_image = tf.image.resize_images(
    image, 
    [225, 422], 
    method=tf.image.ResizeMethod.BILINEAR
)
```

```python
gray_resized_image = tf.image.rgb_to_grayscale(resized_image)
```

```python
init_op = tf.global_variables_initializer()
with tf.Session() as s:
    s.run(init_op)
    img = s.run(gray_resized_image)
    print(img.shape)
    plt.figure()
    plt.imshow(np.squeeze(img,2), cmap='gray')
    plt.show()
```
